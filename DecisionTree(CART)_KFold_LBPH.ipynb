```python
import os
import numpy as np
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
from skimage import io, feature
from skimage.transform import resize
import time  
```

```python
path = "/Users/rizqi/Documents/Jupyter Notebook/Class Classification/archive"
```

```python
def load_images(folder_path, method='lbph', target_size=(700, 1000), R=2, P=16):
    images = []
    labels = []
    filenames = []

    for label, class_name in enumerate(['email', 'resume', 'scientific_publication']):
        class_path = os.path.join(folder_path, class_name)
        for filename in os.listdir(class_path):
            if filename.endswith(".png"):
                img_path = os.path.join(class_path, filename)
                image = io.imread(img_path)

                feature_image = feature.local_binary_pattern(image, P=P, R=R, method='uniform')
                feature_image = np.histogram(feature_image.ravel(), bins=np.arange(0, 10), range=(0, 10))[0]

                feature_image = feature_image / feature_image.max()

                resized_image = resize(feature_image, target_size, anti_aliasing=True)

                images.append(resized_image.flatten())
                labels.append(label)
                filenames.append(filename)

    return images, labels, filenames
```

```python
X, y, filenames = load_images(path)

kfold = KFold(n_splits=5, shuffle=True, random_state=None)

accuracies = []
specificities = []
precisions = []
recalls = []
f1_scores = []

elapsed_times = []
```

```python
for fold_idx, (train_index, test_index) in enumerate(kfold.split(X, y)): 
    start_time = time.time()
    
    X_train, X_test = np.array([X[i] for i in train_index]), np.array([X[i] for i in test_index]) 
    y_train, y_test = np.array([y[i] for i in train_index]), np.array([y[i] for i in test_index]) 
    filenames_test = [filenames[i] for i in test_index]
    
    print(f"Fold {fold_idx + 1}:") 
    print(f"Data Train: {len(X_train)}") 
    print(f"Data Test: {len(X_test)}")

    model = DecisionTreeClassifier(max_depth=15, min_samples_split=5, min_samples_leaf=3, random_state=None) 
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred) 
    print(f"Accuracy: {accuracy}")

    print(f"\nTest Data Files and Paths:") 
    for filename_test in filenames_test:
        img_path_test = os.path.join(path, filename_test) 
        print(f"File: {filename_test}, Path: {img_path_test}\n")

    specificity = cm[1, 1] / (cm[1, 0] + cm[1, 1])
    
    report = classification_report(y_test, y_pred, target_names=['email', 'resume', 'scientific_publication']) 
    print(report)
    print("="*102)

    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted') 
    accuracies.append(accuracy)
    specificities.append(specificity)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1_score)
    
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['email', 'resume', 'scientific_publication'],
                yticklabels=['email', 'resume', 'scientific_publication']) 
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted') 
    plt.ylabel('Actual') 
    plt.show()
    
    plt.figure(figsize=(20, 10))
    plot_tree(model, filled=True, feature_names=[f'pixel_{i}' for i in range(X_train.shape[1])], class_names=['email', 'resume', 'scientific_publication']) 
    plt.show()

    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Elapsed Time: {elapsed_time} seconds")
    elapsed_times.append(elapsed_time)
    print("\n" * 3)

```

```python
print("\nAverage Metrics Across Folds:")
print(f"Mean Accuracy: {np.mean(accuracies)}")
print(f"Mean Specificity: {np.mean(specificities)}")
print(f"Mean Precision: {np.mean(precisions)}")
print(f"Mean Recall: {np.mean(recalls)}")
print(f"Mean F1-score: {np.mean(f1_scores)}")

print("\n" * 1)

print(f"Mean elapsed time: {np.mean(elapsed_times)} seconds")
total_time = sum(elapsed_times)
print(f"Total elapsed time for All folds: {total_time} seconds")
```
